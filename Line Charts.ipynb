```python
# Display pycharts in Jupyter Notebooks inline
%matplotlib inline

# Adding dependency
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Skill Drill
# Using the object-oriented approach, make the following changes to your line chart:
# Change the line color to green.
# Change the marker to a diamond.
# Change the line width to 2 points.
# Add a legend for the city of Boston.
# Add a title and axes labels.
# Add a y-axis limit.
# Add grid lines.

fig = plt.figure() 
ax = fig.add_subplot()
ax.plot(x_axis, y_axis, label= 'Boston', color='green', marker='d', linestyle='dashed')
#ax.plot(x_axis, y_axis, label='Boston', color="green")
ax.set_xlim('Jan', 'Dec')
ax.set_ylim(0, 45)
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.set_title("PyBer Fare by Month")
ax.legend()
ax.grid()
#ax.plot(x_axis,y_axis,'gd-')
```
