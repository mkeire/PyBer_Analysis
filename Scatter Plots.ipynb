```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
plt.scatter(x_axis, y_axis, label = 'Chicago', c = 'red')
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
plt.scatter(x_axis, y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, label='Boston', s=[i * 5 for i in y_axis], c='skyblue', norm=.2, linewidths=2, edgecolors='black')
ax.set_xlim(0,45)
ax.set_ylabel('Date')
ax.set_xlabel('Fare ($)')
ax.set_title("PyBer Fare by Month")
ax.legend(loc=3)
ax.invert_yaxis()

# s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, 
# linewidths=None, verts=None, edgecolors=None, *, plotnonfinite=False, data=None,
```
