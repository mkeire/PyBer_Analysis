```python
# Matplotlib inline magic command
%matplotlib inline

# Adding dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as sts
import matplotlib.style as style


# Loading data files
city_data_to_load = 'Resources/city_data.csv'
ride_data_to_load = 'Resources/ride_data.csv'

# Read in data files
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

```python
# Data Check
city_data_df.count()
city_data_df.isnull().sum()
city_data_df.dtypes
city_data_df['type'].unique()
sum(city_data_df['type']=='Urban')
sum(city_data_df['type']=='Rural')
ride_data_df.count()
ride_data_df.isnull().sum()
ride_data_df.dtypes
```

```python
# Merging Dataframes
pyber_data_df = pd.merge(ride_data_df, city_data_df, how='left', on=['city', 'city'])
# pyber_data_df.head()
```

```python
# Creating dataframes by city type
urban_cities_df = pyber_data_df[pyber_data_df['type'] == 'Urban']
suburban_cities_df = pyber_data_df[pyber_data_df['type'] == 'Suburban']
rural_cities_df = pyber_data_df[pyber_data_df['type'] == 'Rural']
```

```python
# Summing rides by city type
urban_ride_count = urban_cities_df.groupby(['city']).count()['ride_id']
suburban_ride_count = suburban_cities_df.groupby(['city']).count()['ride_id']
rural_ride_count = rural_cities_df.groupby(['city']).count()['ride_id']
```

```python
# Calculating the mean for fares by city type
urban_avg_fare = urban_cities_df.groupby('city').mean()['fare']
suburban_avg_fare = suburban_cities_df.groupby(['city']).mean()['fare']
rural_avg_fare = rural_cities_df.groupby(['city']).mean()['fare']
urban_avg_fare
```

```python
# Calculating the mean number of drivers by city type
urban_driver_count = urban_cities_df.groupby(['city']).mean()['driver_count']
suburban_driver_count = suburban_cities_df.groupby(['city']).mean()['driver_count']
rural_driver_count = rural_cities_df.groupby(['city']).mean()['driver_count']
```

```python
# Create a scatter plot of the urban total number of riders per city by the mean fare in dollars
fig, ax = plt.subplots()
mean_fare_urban_scatter = ax.scatter(urban_ride_count, urban_avg_fare,
                               s=10*urban_driver_count, label='Urban',
                               c='coral', alpha=0.8, linewidths=1,
                               edgecolors='black')
ax.set_xlabel('Total Number of Riders (Per City)')
ax.set_ylabel('Average Fare ($)')
ax.set_title('PyBer Ride-Sharing Data (2019)')
ax.legend()
ax.grid(True)
plt.show()
```

```python
# Pyber Challenge Part 1

# Code Templates:
# https://www.geeksforgeeks.org/creating-a-dataframe-from-pandas-series/
# https://pandas.pydata.org/pandas-docs/stable/user_guide/style.html

# pyber_data_df

# Sum drivers by city type
sum_drivers_by_type = city_data_df.groupby(["type"]).sum()["driver_count"]

# Sum Rides by city type
sum_rides_by_type = pyber_data_df.groupby(["type"]).count()["ride_id"]

# Calculate avearge fare per ride
avg_fare_per_ride = pyber_data_df.groupby('type').mean()['fare']

# Calculate average fare per driver
avg_fare_per_driver = sum_fares_by_type / sum_drivers_by_type

# Creating summary data frame: total rides, total drivers, total fares, average fare per ride, average fare per driver
summary = {'Total Rides': sum_rides_by_type, 'Total Drivers': sum_drivers_by_type,
           'Total Fares': sum_fares_by_type, 'Average Fare Per Ride': avg_fare_per_ride,
           'Average Fare Per Driver': (avg_fare_per_driver)}
summary_df = pd.DataFrame(summary)

# Formatting
summary_df.style.format({'Total Rides': '{:,.0f}', 'Total Drivers': '{:,.0f}', 'Total Fares': '${:,.2f}',
                        'Average Fare Per Ride': '${:,.2f}', 'Average Fare Per Driver': '${:,.2f}'})
```

```python
# Pyber Challenge Part 2

# Rename dataframe columns
pyber_data_df_renamed = pyber_data_df.rename(columns={'city': 'City', 'date':'Date',
                                              'fare':'Fare', 'ride_id': 'Ride Id',
                                              'driver_count': 'No. Drivers',
                                              'type':'City Type'})
# pyber_data_df_renamed
```

```python
# Set Date to datime64 data type !Outputs series!
pyber_data_date_to_datetime = pyber_data_df_renamed['Date'].astype('datetime64[ns]')

# Change Series to DataFrame
date_frame = pyber_data_date_to_datetime.to_frame()

# Merge corrected date data type
dates_merged = pd.merge(pyber_data_df_renamed, date_frame, left_index=True, right_index=True)
dates_merged
```

```python
# Drop dates with incorrect data type, Ride Id, and No. Drivers
pyber_new_dates = dates_merged.drop(columns=['Date_x', 'Ride Id', 'No. Drivers'])

# Rename dates with correct data type
pyber_new_dates_df = pyber_new_dates.rename(columns={'Date_y': 'Date'})

# Set index to date
pyber_date_index_df = pyber_new_dates_df.set_index('Date')
pyber_date_index_df
```

```python
# Calculate the fare total by the type of city and date
pyber_grouped = pyber_date_index_df.groupby(['City Type', 'Date']).sum()

# Reset index
pyber_reset = pyber_grouped.reset_index()
pyber_set = pyber_reset.set_index('Date')
pyber_set
```

```python
# Create pivot table
pyber_pivot = pyber_set.pivot_table(index=pyber_set.index, values='Fare', columns='City Type')
# print(pyber_pivot)

# Convert pivot table to dataframe
pyber_pivoted_df = pd.DataFrame(pyber_pivot)

# Filter for date range using loc
pivot_table_new = pyber_pivoted_df.loc['2019-01-01':'2019-04-28']

# Resample weekly
pyber_resample = pivot_table_new.resample('W').sum()

# pyber_pivoted_df
# pivot_table_new
pyber_resample
```

```python
# # FiveThirtyEight style: https://www.dataquest.io/blog/making-538-plots/
plt.style.use('fivethirtyeight')
pyber_resample.plot(figsize = (20,10))
plt.ylabel('Fare: $ USD)')
plt.xlabel('Month')
ax1.set_title('Total Fare by City Type')
ax1.legend(['Rural', 'Suburban', 'Urban'])
plt.tight_layout()

# Save figure.
plt.savefig("analysis/Fig10.png", bbox_inches='tight')

plt.show()
```
